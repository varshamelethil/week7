## 3.Carry out correlation analysis on the dataset and find the correlation factor between        a. ph and Chloramines b. ph and Conductivity What do we infer from this correlation value? 

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

##### a. ph and Chloramines

```python
x=data['ph']
y= data['Chloramines']
```

```python
#calculating Pearson's correlation
corr,_=pearsonr(x,y)
print('pearsons correlation:%.3f'% corr)
```

```python
# calculating spearman's correlation
corr,_=spearmanr(x,y)
print('spearmans correlation:%.3f'%corr)
```

```python
y=data[['ph','Chloramines']]
y.corr()
```

```python
data.plot(kind="scatter",x="ph",y="Chloramines")
```

```python
sns.heatmap(y.corr(),xticklabels=y.corr().columns.values,yticklabels=y.corr().columns.values,annot=True)
plt.show()
```


* Both ph and chloramines having same correlation coeffecient value of-0.025 and correlation values are low.
* Negatively correlated
* When pH increases ,the chloramine  value is decreases vise versa.



####  b. ph and Conductivity 

```python
x=data['ph']
y= data['Conductivity']
```

```python
#calculating Pearson's correlation
corr,_=pearsonr(x,y)
print('pearsons correlation:%.3f'% corr)
```

```python
# calculating spearman's correlation
corr,_=spearmanr(x,y)
print('spearmans correlation:%.3f'%corr)
```

```python
x=data[['ph','Conductivity']]
x.corr()
```

```python
data.plot(kind="scatter",x="ph",y="Conductivity",color='green')
```

```python
sns.heatmap(x.corr(),xticklabels=x.corr().columns.values,yticklabels=x.corr().columns.values,annot=True)
plt.show()
```

* Both pH and Conductivity having same correlation coeffecient value of 0.014.
* Positively correlated.
* when the  pH increases , then conductivity also increases.
